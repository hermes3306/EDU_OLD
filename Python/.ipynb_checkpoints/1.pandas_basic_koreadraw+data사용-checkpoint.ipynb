```python
data.loc[data['population'] == max_pop]
```

```python
def get_max_pop_df(sido_name):
    max_pop = data.loc[data['광역시도'] == sido_name,'population'].max()
    return data.loc[data['population'] == max_pop,['광역시도','행정구역','population']]
```

```python
get_max_pop_df('경상북도')
```

```python
#새로운 DataFrame을 생성
max_pop_df = \
pd.DataFrame(columns=['광역시도','행정구역','population'])

for sido_name in data['광역시도'].unique():
    max_data = get_max_pop_df(sido_name)
    max_pop_df = max_pop_df.append(max_data)
    
max_pop_df
```

```python
# max_pop_df 의 index값을 reset
max_pop_df = max_pop_df.reset_index()
#index 컬럼을 삭제
del max_pop_df['index']
max_pop_df
```

```python
max_pop_df.sort_values(by='population',ascending=False)
```

```python
data['면적'].corr(data['population'])
```

```python
# Group by 기능 사용
# 광역시도별 인구수의 합계
#1. Series 의 groupby()


data['population'].groupby(data['광역시도']).sum().sort_values(ascending=False)
```

```python
#2. DataFrame의 groupby()
data.groupby('광역시도')['population'].sum().sort_values(ascending=False)
```

```python
#광역시도별, 행정구역별 인구수의 합계
data.groupby(['광역시도','행정구역'])['population'].sum().sort_values(ascending=False)
```

```python
#2. DataFrame의 groupby()
data2 = data.groupby(['광역시도','행정구역'])['population'].sum().sort_values(ascending=False)
data2.shape
```

```python
#3. save
data2.to_excel('data/population.xls',sheet_name='pop')
```

```python
data2_seoul = data2.loc['서울특별시'].sort_values(ascending=False)
data2_seoul.to_excel('data/seoulpop.xls')
data2_seoul
```

```python
#Figure, Axes를 생성
figure, (ax1,ax2) = plt.subplots(nrows=2,ncols=1)
figure.set_size_inches(18,8)
sns.barplot(data=seoul_df, x='행정구역',y='population', ax=ax1)
sns.barplot(data=seoul_df, x='행정구역',y='면적', ax=ax2)
```

```python
df_seoul = seoul_df.sort_values(['population'],ascending=False).reset_index(drop=True)
df_seoul.head()
figure, ax1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(18,12)
sns.barplot(data=df_seoul, x="행정구역", y="population", ax=ax1)
ax1.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
ax1.set( ylabel='인구수')
for item in ax1.get_xticklabels(): 
    item.set_rotation(90)
for i, v in enumerate(df_seoul["population"].iteritems()):        
    ax1.text(i ,v[1], "{:,}".format(v[1]), color='m', va ='bottom', rotation=45)
plt.tight_layout()
```
